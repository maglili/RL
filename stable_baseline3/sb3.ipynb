```python
import gym
import time
from stable_baselines3 import A2C
import numpy as np
import matplotlib.pyplot as plt

env = gym.make("CartPole-v1")

model = A2C("MlpPolicy", env, verbose=1)
model.learn(total_timesteps=100000)
```

```python
episodes = 5
epi_returns = []
for epi in range(5):
    obs = env.reset()
    epi_return = 0
    count = 0
    while True:
        action, _state = model.predict(obs, deterministic=True)
        obs, reward, done, info = env.step(action)
        epi_return += reward
        env.render()
        time.sleep(0.01)
        count += 1   
        if done:
            obs = env.reset()
            epi_returns.append(epi_return)
            print("total step: {:3} return:{:4}".format(count, epi_return))
            break
env.close()
```

```python
print("mean returns:", np.mean(epi_returns))
```

```python
plt.bar([i for i in range(5)],epi_returns)
plt.xlabel('episode')
plt.ylabel('return')
plt.title('Return vs episodes')
plt.show()
```
